```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
coords = np.load('./../Dataset/CaltechFaceSet/raw/coords.npy')
```

```python
for iFace in range(450):
    print(iFace)
    im = misc.imread('./../Dataset/CaltechFaceSet/raw/image_%04d.jpg'%(iFace+1))
    coord = coords[:,iFace]

    x = [int(min(coord[0],coord[2])),int(max(coord[4],coord[6]))]
    y = [int(min(coord[1],coord[3])),int(max(coord[5],coord[7]))]
    
    dW = -10
    
    x[0] = max(0,x[0]-dW)
    x[1] = min(im.shape[1],x[1]+dW)
    y[0] = max(0,y[0]-dW)
    y[1] = min(im.shape[0],y[1]+dW)
    
#     width = x[1]-x[0]+50
#     height = y[1]-y[0]+50
#     L = max(width, height)
    
#     xmid = (x[1]+x[0])//2
#     xl = max(0,xmid -L//2)
#     xr = min(im.shape[1]-1,xl + L)
#     x = [xl,xr]
    
#     ymid = (y[1]+y[0])//2
#     yt = max(0,ymid -L//2)
#     yb = min(im.shape[0]-1,yt + L)
#     y = [yt,yb]
        
    face = im[y[0]:y[1],x[0]:x[1]]
    misc.imsave('./../Dataset/CaltechFaceSet/processed/cf%04d.png' % iFace, face)
#[x_bot_left y_bot_left x_top_left y_top_left x_top_right y_top_right x_bot_right y_bot_right]
```
